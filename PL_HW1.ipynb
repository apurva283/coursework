**Part 1**




 Implement a function gcd, which takes two integers as its arguments and returns their GCD. You may assume that both inputs are postive integers (i.e. m, n € Z and m, n > 0)




```python
#1.1
def gcd(m, n):
    while n != 0:
        a = m % n
        m = n
        n = a
    return m

#example:
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
result = gcd(a, b)
print("The GCD of", a, "and", b, "is", result)
```

What does your function do if one or both of the arguments are negative?

```python
#1.2
gcd(-12, 8)
```

```python
#1.2
gcd(-12, -8)
```

ans:If negative numbers are used, the mathematical GCD remains positive, but the program may behave differently unless we convert numbers using absolute value.(modified code is given below)

```python
def gcd(m, n):
    m = abs(m)
    n = abs(n)

    while n != 0:
        r = m % n
        m = n
        n = r

    return m
print(gcd(-48,-18))
```

**Problem 2: Approximating Euler's number e**


**Part 1**

 Define a function called euler_limit that takes as an argument an integer n, and returns a float that approximates e by taking x = n in Equation (2). You may assume that the input to your function will be a positive integer (of course, your function will still run just fine if n is, say, a positive float, but we will only use integer values in what follows).



```python
#2.1
def euler_limit(n):
    return (1 + 1/n) ** n
print(euler_limit(67))
```

**Part 3**


Define a function called euler_sum that takes a single non-negative integer argument n, and returns an approximation to e based on the first n terms of the sum in Equation (1).

```python
#2.3
def factorial(k):
    result = 1
    for i in range(1, k+1):
        result = result * i
    return result
def euler_sum(n):

    if n == 0:
        return 0

    total = 0

    for k in range(n):
        total = total + 1 / factorial(k)

    return total
print(euler_sum(6))
```

Part 4


Define a function called euler_approx that takes a single argument, a positive float epsilon, and uses the sum in Equation (1) to obtain an approximation of e that is within epsilon of the true value of e.


```python
#2.4
import math

def euler_approx(epsilon):

    true_e = math.exp(1)

    total = 0
    k = 0

    while abs(total - true_e) > epsilon:
        total = total + 1 / math.factorial(k)
        k = k + 1

    return total
print(euler_approx(0.0001))
```

**Problem 3: Testing properties of integers**

Part 1


Write a function is_power_of_2 that takes a positive integer as its only argument and returns a Boolean indicating whether or not the input is a power of 2.


```python
#3.1
def is_power_of_2(n):

    if n == 1:
        return True

    if n % 2 != 0:
        return False

    return is_power_of_2(n // 2)
print(is_power_of_2(8))
print(is_power_of_2(11))
```

Part 2


Generalize your previous solution to a function is_power that takes two positive integers as its arguments, b and n, in that order, and returns a Boolean. is_power (b,n) should return True if n is a power of b and False otherwise.

```python
#3.2
def is_power(b, n):

    if n == 1:
        return True

    if n % b != 0:
        return False

    return is_power(b, n // b)
print(is_power(2, 16))
print(is_power(3, 9))
print(is_power(8, 20))
```
